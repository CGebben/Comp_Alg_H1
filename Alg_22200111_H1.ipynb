```python
def fact_iterative(n):
    '''Function to find factorial of argument n using iterative method.'''
    
    #Initializes f as 1 for a baseline.
    f = 1

    #Check if n is negative. Print statement stating n must be positive.
    if n < 0:
        return 'Integer n must be positive.'
    
    #Check if n is zero. Returns f as 1 since factorial of 0 is 1.
    elif n == 0:
        return f
    
    #For positive integers.
    #For loop for i from 1 to n.
    #Variable f becomes f multiplied by i.
    #After loop ends returns f.
    else:
        for i in range(1, n + 1):
            f *= i
        return f
```

```python
def fact_non_tail_recursive(n):
    '''Function to find factorial of argument n using non-tail-recursive method.'''

    # Check if n is negative. Print statement stating n must be positive.
    if n < 0:
        return 'Integer n must be positive.'
    
    # Check if n is zero. Returns 1 since factorial of 0 is 1.
    if n == 0:
        return 1
    
    # Multiplies n by a recursive call of the function with n - 1 as the argument.
    else:
        return n * fact_non_tail_recursive(n - 1)
```

```python
def fact_tail_recursive(n, a = 1):
    '''Function to find factorial of argument n using tail-recursive method.'''
    
    # Check if n is negative. Print statement stating n must be positive.
    if n < 0:
        return 'Integer n must be positive.'
    
    # Return base case a when n reaches 0
    elif n == 0:
        return a
    
    # Call the function recursively with n - 1 and a * n as arguments.
    else:
        return fact_tail_recursive(n - 1, a * n)
```

```python
def linear_search(array, target):
    """
    Function to perform a linear search on an array and to return a print statement.
    """

    #For loop through array.
    #If index of array is equal to the target number, then prints statement with the index number.
    for i in range(len(array)):
        if array[i] == target:
            return True
        
    #Print statement if number is not found in array.
    return False
```

```python
def binary_search_iterative(array, target):
    """
    Function to perform an iterative binary search on a sorted array and return a print statement.
    """
    
    # Set the left and right indices of the array.
    left = 0
    right = len(array) - 1
    
    # While the left index is less than or equal to the right index:
    while left <= right:
        
        # Find midpoint of array.
        mid = (left + right) // 2
        
        # If the value at the midpoint index is equal to the target, return the index.
        if array[mid] == target:
            return True
        
        # If the value at the midpoint is greater than target, search left half of the array.
        elif array[mid] > target:
            right = mid - 1
        
        # If the value at the midpoint is less than target, search right half of the array.
        else:
            left = mid + 1
    
    # Return print statement if the target is not found in the array.
    return False
```

```python
#Create function to create array of random integers of varying lengths.

def create_array(n):
    '''Function to create an array of random integers of n length.'''

    array = []

    for i in range(n):
        array.append(random.randint(1, 500000))

    return array
```

```python
#Create arrays of multiple lengths to test with search algorithms.

array_5000 = sorted(create_array(5000))
array_7500 = sorted(create_array(7500))
array_10000 = sorted(create_array(10000))
array_15000 = sorted(create_array(15000))
array_20000 = sorted(create_array(20000))
array_30000 = sorted(create_array(30000))
array_50000 = sorted(create_array(50000))
array_75000 = sorted(create_array(75000))
array_100000 = sorted(create_array(1000))
array_125000 = sorted(create_array(125000))
array_150000 = sorted(create_array(150000))
array_175000 = sorted(create_array(175000))
array_200000 = sorted(create_array(200000))
array_225000 = sorted(create_array(225000))
array_250000 = sorted(create_array(250000))
```

```python
def time_linear_search(array, target):
    '''Function to search an array for a target int using linear_search function. Returns the run_time.'''
    start_time = time.time()
    linear_search(array, target)
    end_time = time.time()
    return round(end_time - start_time, 8)

def time_binary_search(array, target):
    '''Function to search an array for a target int using binary_search function. Returns the run_time.'''
    start_time = time.time()
    binary_search_iterative(array, target)
    end_time = time.time()
    return round(end_time - start_time, 8)

def time_array_linear(array, target):
    '''Function to create an array of 50 runtimes for an array put through the time_linear_search algorithm.'''
    runtimes = []
    for j in range(50):
        time = time_linear_search(array, target)
        runtimes.append(time)
    return runtimes

def time_array_binary(array, target):
    '''Function to create an array of 50 runtimes for an array put through the time_binary_search algorithm.'''
    runtimes = []
    for j in range(50):
        time = time_binary_search(array, target)
        runtimes.append(time)
    return runtimes

```

```python
#Create array to store best, worst, and average times for both linear and binary searches.
linear_search_best = []
linear_search_worst = []
linear_search_avg = []
binary_search_best = []
binary_search_worst = []
binary_search_avg = []

for array in [array_5000, array_7500, array_10000, array_15000, array_20000, array_30000, array_50000, array_75000, array_100000, array_125000, array_150000, array_175000, array_200000, array_225000, array_250000]:
    #Target will be 1300 for all searches.
    target = 1300

    #Store array of runtimes for linear search.
    linear_array = time_array_linear(array, target)

    #For linear_search, append min runtime to best array, max runtime to worst array, and avg runtime to avg array.
    linear_search_best.append(min(linear_array))
    linear_search_worst.append(max(linear_array))
    linear_search_avg.append(sum(linear_array) / len(linear_array))

    #Store array of runtimes for binary search.
    binary_array = time_array_binary(array, target)

    #For binary_search, append min runtime to best array, max runtime to worst array, and avg runtime to avg array.
    binary_search_best.append(min(binary_array))
    binary_search_worst.append(max(binary_array))
    binary_search_avg.append(sum(binary_array) / len(binary_array))

#Print arrays to test.
print(linear_search_best)
print(linear_search_worst)
print(linear_search_avg)
print(binary_search_best)
print(binary_search_worst)
print(binary_search_avg)
```
