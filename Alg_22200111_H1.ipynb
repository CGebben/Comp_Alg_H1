```python
def fact_iterative(n):
    '''Function to find factorial of argument n using iterative method.'''
    
    #Initializes f as 1 for a baseline.
    f = 1

    #Check if n is negative. Print statement stating n must be positive.
    if n < 0:
        return 'Integer n must be positive.'
    
    #Check if n is zero. Returns f as 1 since factorial of 0 is 1.
    elif n == 0:
        return f
    
    #For positive integers.
    #For loop for i from 1 to n.
    #Variable f becomes f multiplied by i.
    #After loop ends returns f.
    else:
        for i in range(1, n + 1):
            f *= i
        return f
```

```python
def fact_non_tail_recursive(n):
    '''Function to find factorial of argument n using non-tail-recursive method.'''

    # Check if n is negative. Print statement stating n must be positive.
    if n < 0:
        return 'Integer n must be positive.'
    
    # Check if n is zero. Returns 1 since factorial of 0 is 1.
    if n == 0:
        return 1
    
    # Multiplies n by a recursive call of the function with n - 1 as the argument.
    else:
        return n * fact_non_tail_recursive(n - 1)
```

```python
def fact_tail_recursive(n, a = 1):
    '''Function to find factorial of argument n using tail-recursive method.'''
    
    # Check if n is negative. Print statement stating n must be positive.
    if n < 0:
        return 'Integer n must be positive.'
    
    # Return base case a when n reaches 0
    elif n == 0:
        return a
    
    # Call the function recursively with n - 1 and a * n as arguments.
    else:
        return fact_tail_recursive(n - 1, a * n)
```

```python
import random

#Initialize array to contain random numbers.
num_array = []

#For loop to iterate through array 100 times and add random numbers.
for i in range(1000):
    num_array.append(random.randint(1, 2000))

#Sorting array in numerical order and storing it in same variable.
num_array = sorted(num_array)

#Print statement test.
print(num_array)
```

```python
import time

#Initialize arrays that will contain runtimes.
fact_it_runtime = []
fact_tail_runtime = []
fact_non_tail_runtime = []

#For loop to iterate through number array to find factorial iteratively.
#Finds start time, runs fact_iterative function and then find end time.
#Subtracts start_time from end_time to find run_time.
#Appends run_time to fact_it_runtime array.
for i in num_array:
    start_time = time.time()
    result = fact_iterative(i)
    end_time = time.time()
    run_time = end_time - start_time
    fact_it_runtime.append(run_time)


#For loop to iterate through number array to find factorial using non-tail recursion.
#Finds start time, runs fact_non_tail_recursive function and then finds end time.
#Subtracts start_time from end_time to find run_time.
#Appends run_time to fact_non_tail_runtime array.
for i in num_array:
    start_time = time.time()
    result = fact_non_tail_recursive(i)
    end_time = time.time()
    run_time = end_time - start_time
    fact_non_tail_runtime.append(run_time)


#For loop to iterate through number array to find factorial using tail recursion.
#Finds start time, runs fact_tail_recursive function and then find end time.
#Subtracts start_time from end_time to find run_time.
#Appends run_time to fact_tail_runtime array.
for i in num_array:
    start_time = time.time()
    result = fact_tail_recursive(i)
    end_time = time.time()
    run_time = end_time - start_time
    fact_tail_runtime.append(run_time)
```

```python
import matplotlib.pyplot as plt

# Plot the running time as a function of n for each running time array.
plt.plot(num_array, fact_it_runtime, label='Iterative')
plt.plot(num_array, fact_non_tail_runtime, label='Non-Tail Recursive')
plt.plot(num_array, fact_tail_runtime, label='Tail Recursive')

# Add title and axis labels to plot.
plt.title('Factorial Function Running Time')
plt.xlabel('n')
plt.ylabel('Running Time (Seconds)')

# Add legend to plot.
plt.legend()

# Show the plot.
plt.show()
```

```python
def linear_search(array, target):
    """
    Function to perform a linear search on an array and to return a print statement.
    """

    #For loop through array.
    #If index of array is equal to the target number, then prints statement with the index number.
    for i in range(len(array)):
        if array[i] == target:
            return print(f"{target} found at index {i}.")
        
    #Print statement if number is not found in array.
    return print(f"{target} is not found in array.")
```

```python
def binary_search_iterative(array, target):
    """
    Function to perform an iterative binary search on a sorted array and return a print statement.
    """
    
    # Set the left and right indices of the array.
    left = 0
    right = len(array) - 1
    
    # While the left index is less than or equal to the right index:
    while left <= right:
        
        # Find midpoint of array.
        mid = (left + right) // 2
        
        # If the value at the midpoint index is equal to the target, return the index.
        if array[mid] == target:
            return print(f"{target} found at index {mid}.")
        
        # If the value at the midpoint is greater than target, search left half of the array.
        elif array[mid] > target:
            right = mid - 1
        
        # If the value at the midpoint is less than target, search right half of the array.
        else:
            left = mid + 1
    
    # Return print statement if the target is not found in the array.
    return print(f"{target} is not found in array.")
```
